```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Train.csv")
df
```

```python
df.info()
```

## Dropping Columns

```python
columns_to_drop=['name','ticket','cabin','boat','body','home.dest']

df.drop(columns=columns_to_drop,inplace=True)
df
```

## Handling Missing data

```python
df['age']=df['age'].fillna(df['age'].mean())
df['fare']=df['fare'].fillna(df['fare'].mean())
df['embarked']=df['embarked'].fillna('S')
df
```

## Encoding String Data to Numeric Data

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['embarked']=le.fit_transform(df['embarked'])
df
```

# Normalization

```python
for f in df.columns:
    if f=='age' or f=='fare':
        df[f]=(df[f]-df[f].mean())/df[f].std()
```

```python
df
```

# Getting X and Y from DF

```python
X=pd.DataFrame([])

for f in df.columns:
    if f!="survived":
        X[f]=df[f].copy()
        
X
```

```python
Y=df['survived'].copy()
Y
```

## Splitting Data into test and train

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)

X_train=X_train.reset_index(drop=True)
X_test=X_test.reset_index(drop=True)
Y_train=Y_train.reset_index(drop=True)
Y_test=Y_test.reset_index(drop=True)

X_train.shape, Y_train.shape
```


## Finding no of trees using RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

acc_list=[]

for i in range(1,61):
    r=RandomForestClassifier(n_estimators=i,criterion="entropy",max_depth=5)
    val=cross_val_score(r,X_train,Y_train,cv=3).mean()
    acc_list.append(val)
    print(i,end=" ")
```

```python
no=np.argmax(acc_list)+1
no
```

```python
r=RandomForestClassifier(n_estimators=no,criterion="entropy",max_depth=5)
r.fit(X_train,Y_train)
r.score(X_train,Y_train),r.score(X_test,Y_test)
```
