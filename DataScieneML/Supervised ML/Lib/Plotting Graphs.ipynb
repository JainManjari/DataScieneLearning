# Customizing Graph

```python
#customizing graphs

x=np.arange(0,5,0.1)
y=x**3
y2=x**2

plt.plot(x,y2,color="red",linewidth=0.55,label="x**2") #if we are using linewidth, then we cant use marker, 
plt.plot(x,y,color="black",linewidth=0.55,label="x**3") #it will have no effect
plt.axis([0,10,0,100]) #four arguments => first two for x starting and ending other two for y
#plt.axis("off") to turn off the axis
plt.ylabel("y")
plt.xlabel("x")
plt.text(2,80,"Text",fontsize=7)
plt.title("Cube Graph")
plt.grid()
plt.legend()
plt.show()
```

# Bubble Graph

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,280,400,450,500]
```

```python
plt.scatter(year,salary,s=population,c='r') #s stands for size, either u can a give a const value or array of the same
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.5,marker="^",edgecolor="black") #s stands for size, either u can a give a const value or array of the same
plt.title("Bubble Chart")
plt.show()
```

```python
uni_years=np.array(data.year)
uni_years=np.unique(uni_years)
jobs=[]
for year in uni_years:
    t=data[data.year==year]
    jobs.append(t.shape[0])
```

```python
location=np.array(data.location)
city=[]
for co in location:
    s=co.split(" ")
    s1=s[0].strip(",")
    if s1=="IN":
        city.append(s[len(s)-2])
```

```python
cities,jobs=np.unique(city,return_counts=True)
jobs_sort_ind=np.argsort(-jobs)
jobs=jobs[jobs_sort_ind]
cities=cities[jobs_sort_ind]
print(cities)
print(jobs)
```

```python
plt.pie(jobs,labels=cities,autopct="%.2f%%",counterclock=False)
plt.axis("equal")
plt.show()
```

```python
pert=np.true_divide(jobs,jobs.sum())*100
pert
```

```python
for i in range(len(cities)):
    print(cities[i],format(pert[i],' .2f'))
```
