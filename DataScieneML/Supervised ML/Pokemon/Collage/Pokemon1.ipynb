# ML On Toy Dataset

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

X,Y=make_blobs(n_samples=400,n_features=2,centers=2)

print(X.shape)
print(Y.shape)
```

```python
test_fruit=np.array([0,10])
plt.style.use("classic")
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(test_fruit[0],test_fruit[1],c="yellow",marker="*")
plt.show()
```

```python
def dist(a1,a2):
    return np.sum((a1-a2)**2)**0.5

def KKN(X,Y,test_fruit,k=5):
    m=X.shape[0]
    vals=[]
    for i in range(m):
        d=dist(test_fruit,X[i])
        vals.append((d,Y[i])) #Storing distance and the label whether the distance is measured from brown or blue
        
    vals=sorted(vals)
    vals= np.array(vals[:k],dtype=int) #take first k shortest distances and their labels
    b=np.unique(vals[:,1],return_counts=True) #it returns the label values nearest to the test fruit
    ind=np.argmax(b[1]) #it returns the index of the max element of the b[1] array which is frequency array of labels
    return b[0][ind] #by getting the index, we can deduce which element has the highest frequency as elements are sorted 

KKN(X,Y,test_fruit,k=100)
    
```
